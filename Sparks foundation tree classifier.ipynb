# Prediction using Decision Tree Algorithm

## Create the decision tree classifier and visualize it graphically

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
pip install scikit-learn
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Create a decision tree classifier
clf = DecisionTreeClassifier()
clf.fit(X, y)
```

```python
# Visualize the decision tree graphically
plt.figure(figsize=(12, 8))
tree.plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```

```python
# Define a new data point (features) for prediction
new_data_point = np.array([[5.1, 3.5, 1.4, 0.2]])  # Replace with your own feature values
```

```python
# Predict the class for the new data point
predicted_class = clf.predict(new_data_point)
```

```python
# Map the class label to the corresponding class name
class_names = iris.target_names
predicted_class_name = class_names[predicted_class[0]]
```

```python
print(f"Predicted Class: {predicted_class_name}")
```
